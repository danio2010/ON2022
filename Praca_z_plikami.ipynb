### Zadanie
Wczytaj plik ```mnist_train_small.csv```. Wypisz jego pięć pierwszych wierszy stawiając na początku numer wiersza, a na końcu wykrzyknik (po każdym wierszu).

Zapisz też to wszystko w nowym pliku txt.

````python
wiersze=[]
with open('sample_data/mnist_train_small.csv') as plik:
  for i in range(5):
    nowy_wiersz=f'{i}: {plik.readline()[:-2]}!'
    wiersze.append(nowy_wiersz)
    print(nowy_wiersz)
with open('sample_data/nowe_wiersze.txt', 'w') as plik:
  for el in wiersze:
    plik.write(el+'\n')
````
